## ジェネレータ　　　メモリ効率のよいイテラブルなオブジェクト

ジェネレータはリストや,タプルのように、for文で利用できるイテラブルなオブジェクト(以下、イテラブル)です。リストやタプルは、すべての要素をメモリ上に保持するため、要素数が増えれば増えるほどメモリ使用量も増える欠点があります。これに対しジェネレータは、次の要素が求められるたびに新たな要素を生成して返せます。つまり、要素数にかかわらずメモリ使用量を地位差kう保てます。

### ジェネレータの具体例

```python
# yieldを含む関数はジェネレータになる
def inf(n):
    while True:
        yield n
```

```python
# Ctrl+Cで中断できる
for i in inf(3):
    print(i)
```

### ジェネレータの実装　　関数のように作成する

```python
def gen_function(n):
    print('start')
    while n:
        print(f'yield: {n}')
        yield n # ここで一時中断される
        n -= 1
```

```python
# 戻り値はジェネレータイテレータ
gen = gen_function(2)
gen
```

```python
# 組み込み関数next()に渡すと
# __next__()が呼ばれる
next(gen)
```
