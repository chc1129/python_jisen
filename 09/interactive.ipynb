## ジェネレータ　　　メモリ効率のよいイテラブルなオブジェクト

ジェネレータはリストや,タプルのように、for文で利用できるイテラブルなオブジェクト(以下、イテラブル)です。リストやタプルは、すべての要素をメモリ上に保持するため、要素数が増えれば増えるほどメモリ使用量も増える欠点があります。これに対しジェネレータは、次の要素が求められるたびに新たな要素を生成して返せます。つまり、要素数にかかわらずメモリ使用量を地位差kう保てます。

### ジェネレータの具体例

```python
# yieldを含む関数はジェネレータになる
def inf(n):
    while True:
        yield n
```

```python
# Ctrl+Cで中断できる
for i in inf(3):
    print(i)
```

### ジェネレータの実例　　　ファイルの内容を変換する

```python
# ファイルの中身を1行ずつ読み込む
def reader(src):
    with open(src) as f:
        for line in f:
            yield line
```

```python
# 行単位で実行する変換処理
def convert(line):
    return line.upper()
```

```python
# 読み込み→変換→書き込みを1行ずつ行う
def writer(dest, reader):
    with open(dest, 'w') as f:
        for line in reader:
            f.write(convert(line))
```

```python
# reader()に存在するファイルのパスを渡す
writer('dest.txt', reader('src.txt'))
```

### functools.lru_cache()　　　関数の結果をキャッシュする関数デコレータ

```python
from functools import lru_cache
from time import sleep

# 最近の呼び出し最大32回分までキャッシュ
@lru_cache(maxsize=32)
def heavy_function(n):
    sleep(3) # 重い処理をシミュレート
    return n + 1
```

### シンプルなデコレータ

```python
# デコレータしたい関数を受け取る
def decol(f):
    print('decol called')
    def wrapper():
        print('before exec')
        v = f() # もとの関数を実行
        print('after exec')
        return v
    return wrapper
```

```python
# デコレータは関数定義時に実行される
@decol
def func():
    print('exec')
    return 1
```

```python
# すでにdeco1(func)の結果へ置き換わっている
func.__name__
```

```python
# func()の呼び出しはwrapper()の呼び出しになる
func() # wrapper()と等価
```

```python
@decol
def func(x, y):
    print('exec')
    return x, y
```

```python
# deco1内のv = f()の行でエラー
func(1, 2)
```

### 引数を受け取る関数のデコレータ

```python
# 引数zを受け取る
def deco3(z):
    # deco2()と同等
    def _deco3(f):
        def wrapper(*args, **kwargs):
            # ここでzを参照できる
            print('before exec', z)
            v = f(*args, **kwargs)
            print('after exec', z)
            return v
        return wrapper
    return _deco3  # デコレータを返す
```

```python
# deco3(z=3)の戻り値がデコレータの実体
# つまりfunc = deco3(z=3)(func)と同等
@deco3(z=3)
def func(x, y):
    print('exec')
    return x, y
```

```python
# zに渡した値は保持されている
func(1, 2)
```

```python
### 複数のデコレータを同時に利用する
```

```python
# 複数のデコレータを利用
@deco3(z=3)
@deco3(z=4)
def func(x, y):
    print('exec')
    return x, y
```

```python
# @deco3(z=4)が適用された結果に
# @deco3(z=3)が適用される
func(1, 2)
```

### functools.wraps()でデコレータの欠点を解消する

```python
from functools import wraps
def deco4(f):
    @wraps(f)  # もとの関数を引数に取るデコレータ
    def wrapper(*args, **kwargs):
        print('before exec')
        v = f(*args, **kwargs)
        print('after exec')
        return v
    return wrapper
```

```python
@deco4
def func():
    """funcです"""
    print('exec')
```

### デコレータの実例　　　処理時間の計測

デコレータの実例として、関数の処理時間を計測してみます。デコレータelapsed_time()を次のように定義します。

```python
from functools import wraps
import time
def elapsed_time(f):
    @wraps(f)
    def wrapper(*args, **kwargs):
        start = time.time()
        v = f(*args, **kwargs)
        print(f"{f.__name__}: {time.time() - start}")
        return v
    return wrapper
```

デコレータelapsed_time()を使うと、関数定義時に@elapsed_timeを付けるだけでその関数の処理時間を計測できます。関数のロジックに変更を加える必要もなく、複数の関数を同時に計測したり、計測対象の関数を変更したりするのも簡単です。

```python
# 0からn-1までの総和を計算する関数
@elapsed_time
def func(n):
    return sum(i for i in range(n))
```

```python
# func()の実行結果を表示
# f-stringで数値のカンマ（,）区切りを指定
print(f'{func(1000000):,}')
```

```python
print(f'{func(10000000):,}')
```

### \_\_enter\_\_()、\_\_exit\_\_()　　　with文の前後に呼ばれるメソッド

```python
# このクラスのインスタンスがコンテキストマネージャー
class ContextManager:
    # 前処理を実装
    def __enter__(self):
        print('__enter__ was called')
    # 後処理を実装
    def __exit__(self, exc_type, exc_value, traceback):
        print('__exit__ was called')
        print(f'{exc_type}')
        print(f'{exc_value}')
        print(f'{traceback}')
```

```python
# withブロックが正常終了の場合は
# __exit__()の引数はすべてNone
with ContextManager():
    print('inside the block')
```

### with文と例外処理

```python
# withブロック内で例外が発生した場合は
# その情報が__exit__()に渡される
with ContextManager():
    1 / 0
```

### asキーワード　　　\_\_enter\_\_()の戻り値を利用する

```python
class ContextManager:
    # 戻り値がasキーワードに渡される
    def __enter__(self):
        return 1
    def __exit__(self, exc_type, exc_value, traceback):
        pass
```

```python
with ContextManager() as f:
    print(f)
```

```python
# asキーワードの省略
with ContextManager():
    pass
```

```python
class Point:
    def __init__(self, **kwargs):
        self.value = kwargs
    def __enter__(self):
        print('__enter__ was called')
        return self.value # as節で渡される
    def __exit__(self, exc_type, exc_value, traceback):
        print('__exit__ was called')
        print(self.value)
```

```python
with Point(x=1, y=2) as p:
    print(p)
    p['z'] = 3
```

### contextlib.contextmanagerでシンプルに実装する

```python
from contextlib import contextmanager
@contextmanager
def point(**kwargs):
    print('__enter__ was called')
    value = kwargs
    try:
        # ここより上が前処理
        # valueをasキーワードに渡す
        yield value
        # ここより下が後処理
    except Exception as e:
        # エラー時はこちらも呼ばれる
        print(e)
        raise
    finally:
        print('__exit__ was called')
        print(value)
```

```python
with point(x=1, y=2) as p:
    print(p)
    p['z'] = 3
```

### \_\_set\_\_()を実装する　　　データデスクリプタ

```python
# __set__()を持つクラスはデータデスクリプタ
class TextField:
    def __set_name__(self, owner, name):
        print(f'__set_name__ was called')
        print(f'{owner}, {name}')
        self.name = name
    def __set__(self, instance, value):
        print('__set__ was called')
        if not isinstance(value, str):
            raise AttributeError('must be str')
        # ドット記法ではなく属性辞書を使って格納
        instance.__dict__[self.name] = value
    def __get__(self, instance, owner):
        print('__get__ was called')
        return instance.__dict__[self.name]
```

```python
class Book:
    title = TextField()
```

```python
book = Book()
```

```python
# 代入時には__set__()が呼ばれる
book.title = 'Python Practice Book'
```

```python
# 取得時には__get__()が呼ばれる
book.title
```

```python
# 別のインスタンスを作成して代入
notebook = Book()
```

```python
notebook.title = 'Notebook'
```

```python
# それぞれデータを保持している
book.title
```

```python
notebook.title
```

```python
# 文字列以外は代入できない
book.title = 123
```

### デスクリプタの実例　　　プロパティのキャッシュ

```python
class LazyProperty:
    def __init__(self, func):
        self.func = func
        self.name = func.__name__
    def __get__(self, instance, owner):
        if not instance:
            # クラス変数としてアクセスされたときの処理
            return self
        # self.funcは関数なので明示的にインスタンスを渡す
        v = self.func(instance)
        instance.__dict__[self.name] = v
        return v
```

```python
TAX_RATE = 1.10
```

```python
class Book:
    def __init__(self, raw_price):
        self.raw_price = raw_price
    @LazyProperty
    def price(self):
        print('calculate the price')
        return int(self.raw_price * TAX_RATE)
```

```python
book = Book(1980)
```

```python
book.price
```

```python
Book.price
```
